# RNN

```python
input = [1,2,3,4]
output = [2,4,6,8]
```

```python
w = 0.0
b = 0.0
lr = 0.01
```

```python
for epoch in range(100):
    total_loss = 0
    for x,y in zip(input, output):
        Y_pred = w*x +b
        error = Y_pred -y
        loss = error**2
        total_loss += loss

        dw = 2* error*x
        db = 2* error

        w = w - lr*dw
        b = b - lr*db
    if epoch % 10 == 0:
        print(f"epoch {epoch} | Loss: {total_loss} | w: {w} |b: {b}")

test_data = 7
prediction = w*test_data + b
print(prediction)
```
