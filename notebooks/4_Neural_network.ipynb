```python
input = [1,2,3,4]
output = [2,4,6,8]
```

```python
list(zip(input, output))
```

```python
w = 0.0
b = 0.0
lr = .01
```

```python
for epoch in range(1000):
    total_loss = 0
    for x,y in zip(input, output):
       y_pred = w * x + b
       error = (y_pred - y)
       loss = error ** 2
       total_loss += loss


       dw = 2*error*x
       db = 2*error

       w = w -lr*dw
       b = b -lr*db

    if epoch % 100 == 0:
        print(f"Epoch {epoch} | loss: {total_loss} | w: {w} | b: {b}")

test_data = 7
prediction = w * test_data + b
print(prediction)

```
